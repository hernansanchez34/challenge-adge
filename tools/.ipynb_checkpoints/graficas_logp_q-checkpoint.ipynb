```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar CSV del grid search
df_grid = pd.read_csv("f1_gridsearch_results.csv")  # ajusta la ruta si es necesario

# Filtrar para p = 1.0 y ordenar por q
f1_vs_q = df_grid[df_grid["p"] == 1.0].sort_values("q").reset_index(drop=True)

# Extraer log₂(q) y Macro-F1
q_values = f1_vs_q["q"].values
log2_q = np.log2(q_values)
macro_f1_q = f1_vs_q["Macro-F1"].values

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(log2_q, macro_f1_q, marker='o', linestyle='-', color='blue')
plt.title("Macro-F1 vs log₂(q) en BlogCatalog (p=1)")
plt.xlabel("log₂(q)")
plt.ylabel("Macro-F1")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filtrar para q = 1.0 y ordenar por p
f1_vs_p = df_grid[df_grid["q"] == 1.0].sort_values("p").reset_index(drop=True)

# Extraer log₂(p) y Macro-F1
p_values = f1_vs_p["p"].values
log2_p = np.log2(p_values)
macro_f1_p = f1_vs_p["Macro-F1"].values

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(log2_p, macro_f1_p, marker='o', linestyle='-', color='purple')
plt.title("Macro-F1 vs log₂(p) en BlogCatalog (q=1)")
plt.xlabel("log₂(p)")
plt.ylabel("Macro-F1")
plt.grid(True)
plt.tight_layout()
plt.show()
```
