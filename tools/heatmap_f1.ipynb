```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Cargar los resultados desde el CSV
file_path = Path("f1_gridsearch_results.csv")
df = pd.read_csv(file_path)

# Crear tablas pivote para Macro-F1 y Micro-F1
pivot_macro = df.pivot(index="p", columns="q", values="Macro-F1")
pivot_micro = df.pivot(index="p", columns="q", values="Micro-F1")

# Graficar heatmap Macro-F1
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_macro, annot=True, fmt=".3f", cmap="YlGnBu")
plt.title("Macro-F1 segn combinacin de p y q")
plt.xlabel("q")
plt.ylabel("p")
plt.tight_layout()
plt.show()

# Graficar heatmap Micro-F1
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_micro, annot=True, fmt=".3f", cmap="YlOrBr")
plt.title("Micro-F1 segn combinacin de p y q")
plt.xlabel("q")
plt.ylabel("p")
plt.tight_layout()
plt.show()
```
